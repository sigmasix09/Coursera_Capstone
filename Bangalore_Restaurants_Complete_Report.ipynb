### Methodology

#### In this project we will focus on the restaurant cuisine distribution and density across the Bangalore city only. The area of focus is going to be 15 km from the city center.

#### Step 1: We have collected the data of the restaurant in the city using four square API.
#### Step 2: In this step, the relevant information is kept for further use and converted to a pandas dataframe.
#### Step 3: The data is refined on the basis of uniqueness and is grouped based on the restaurant cuisine.
#### Step 4: Analysis begins and data is divided into clusters, and visualizing the results.
#### Step 5: Discussing the results and findings from the analysis for various stakeholders.
#### Step 6: Finally, conclusion is made for the whole effort.

```python
print('There are {} uniques Cuisine.'.format(len(city_venues['Cuisine'].unique())))

# one hot encoding
city_venues_onehot = pd.get_dummies(city_venues[['Cuisine']], prefix="", prefix_sep="")

# Adding city columns at the end
city_venues_onehot['Cuisine'] = city_venues['Cuisine']

# Putting city column at 1st
fixed_columns = [city_venues_onehot.columns[-1]] + list(city_venues_onehot.columns[:-1])
city_venues_onehot = city_venues_onehot[fixed_columns]

city_venues_onehot.sort_values('Cuisine', axis = 0, ascending = True, inplace = True)
city_venues_onehot.reset_index(drop = True, inplace = True)
city_venues_onehot
```

```python
# Grouping rows by Cuisine and taking mean of frequency of occurence of each category
city_venues_grouped = city_venues_onehot.groupby('Cuisine').mean().reset_index()
city_venues_grouped
```

### Clustering the data for analysis

```python
# set number of clusters
kclusters = 32

city_venues_grouped_clustering = city_venues_grouped.drop('Cuisine', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0, init="k-means++", n_init=12).fit(city_venues_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[:]
```

```python
# To drop the ClusterLabel column in case user changes the kcluster value in previous block
city_venues_grouped.drop(columns='ClusterLabels', inplace = True, axis=1)
```

```python
# add new column called ClusterLabels to the dataframe
city_venues_grouped.insert(0, 'ClusterLabels', kmeans.labels_)
```

```python
city_venues_grouped.head()
```

```python
# Merging two dataframes to add ClusterLabels and various cuisines
city_venues_merged = city_venues
city_venues_merged = city_venues_merged.join(city_venues_grouped.set_index('Cuisine'), on='Cuisine', how = 'inner')
```

```python
city_venues_merged.head()
```

### Marking the clusters on the Bangalore city map

```python
# create map for all clusters
map_bangalore_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map for all clusters
markers_colors = []
for lat, lon, poi, cluster in zip(city_venues_merged['Latitude'], city_venues_merged['Longitude'], city_venues_merged['Cuisine'], city_venues_merged['ClusterLabels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_bangalore_clusters)
    
map_bangalore_clusters
```

```python
# Change the cluster_label_val to see the specific cluster
cluster_label_val = 30
city_venues_merged.loc[city_venues_merged['ClusterLabels'] == cluster_label_val, city_venues_merged.columns[[1] + list(range(5, city_venues_merged.shape[1]))]]
```

### Showing for cluster with more than 4 outlets in the city

```python
# add markers to the map for cluster n
map_bangalore_Mclusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# Uncomment the below codes as per the requirement.
# 1. More Dominating: List of cluster dominating the market
# 2. Less Dominating: List of cluster not dominating the market
more_dominating = []
less_dominating = []

for item in range(kclusters):
    if len(city_venues_merged.loc[city_venues_merged['ClusterLabels'] == item]) <= 4:
        less_dominating.append(item)
    else:
        more_dominating.append(item)

print('Less Dominating clusters : ', less_dominating)
print('More Dominating clusters : ', more_dominating)

# add markers to the map for all clusters
# Switch between less_dominating and more_dominating varibales in beolow code lines as per requirement
markers_colors = []
for lat, lon, poi, cluster in zip(city_venues_merged['Latitude'], city_venues_merged['Longitude'], city_venues_merged['Cuisine'], city_venues_merged['ClusterLabels']):
    if cluster in more_dominating:
        label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color=rainbow[cluster-1],
            fill=True,
            fill_color=rainbow[cluster-1],
            fill_opacity=0.7).add_to(map_bangalore_Mclusters)
    
map_bangalore_Mclusters
```

### Results and Discussions

#### The analysis shows that Bangalore City has vast variety of restaurants and cuisines offered. The 15 km radius from the Bangalore city railway station has been the area of focus in the study. We can see the western and southern regions of the city have higher density of restaurants compared to northern and eastern regions. The key clusters to observe are more dominating clusters, these cluster dominate the city food market and have atleast 5 outlets. Whereas, less dominating cluster  have atmost 4 outlets in the city. 

#### Also, we can observe that the restaurants with cuisines in less dominating clusters are more concentrated in the Central, West and South-Western regions with very less number of outlets in the city per cuisine. Whereas, restaurants with dominating cuisines are well distributed in the city with a very few outliers in the North and East.

#### We can argue that less dominant cluster have good potential to grow in the city and have more regions to cover as presently they are more established in Central and North-Western regions. The enterpreneurs who wish to start restaurants with these cuisines can take good oportunity this and they have significant audience from students to professionals from all over the globe residing in Bangalore.

#### On the other hand, dominant cuisines which cover most of the market share serve the local population and is well scattered. This covers mostly Indian cuisine, cafe and fast food cuisines, Italian being an exception which seems to have established it self with the local cuisine.
